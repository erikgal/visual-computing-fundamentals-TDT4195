```python
import skimage
import skimage.io
import skimage.transform
from skimage.transform import resize
import cv2
import os
import numpy as np
import utils
import matplotlib.pyplot as plt
```

```python
def magnitude(fft_im):
    real = fft_im.real
    imag = fft_im.imag
    return np.sqrt(real**2 + imag**2)
```

```python
if __name__ == "__main__":
    # DO NOT CHANGE
    impath = os.path.join("images", "noisy_moon.png")
    im = utils.read_im(impath)

    ### START YOUR CODE HERE ### (You can change anything inside this block)
    sobel_vertical = np.array([
        [ 1,  2,  1],
        [ 0,  0,  0],
        [-1, -2, -1]
    ])
    conv_im = cv2.filter2D(im, -1, sobel_vertical)
    
    fft = np.fft.fft2(im)
    fft_im = np.fft.fftshift(fft)
    fft_im = np.log(magnitude(fft_im)+1)
    
    fft_kernel = np.fft.fft2(sobel_vertical)
    fft_kernel_shift =  np.fft.fftshift(fft_kernel)
    fft_kernel_shift = np.log(magnitude(fft_kernel_shift)+1)
    
    
    resized_kernel = resize(fft_kernel_shift, fft_im.shape)
    filtered_fft = fft_im * resized_kernel 
    im_filtered = np.fft.ifft2(conv_im).real

    
    # Use plt.subplot to place two or more images beside eachother
    plt.figure(figsize=(20, 4))
    # plt.subplot(num_rows, num_cols, position (1-indexed))
    plt.subplot(1, 5, 1)
    plt.imshow(im, cmap="gray")
    plt.subplot(1, 5, 2) 
    # Visualize FFT
    plt.imshow(fft_im, cmap="gray")
    plt.subplot(1, 5, 3)  
    # Visualize FFT kernel
    plt.imshow(fft_kernel_shift, cmap="gray")
    plt.subplot(1, 5, 4)  
    # Visualize filtered FFT image
    plt.imshow(filtered_fft, cmap="gray")
    plt.subplot(1, 5, 5) 
    # Visualize filtered spatial image
    plt.imshow(conv_im, cmap="gray")

    ### END YOUR CODE HERE ###
    utils.save_im("moon_filtered.png", utils.normalize(im_filtered))
```
